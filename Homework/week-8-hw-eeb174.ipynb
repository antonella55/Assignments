```python
import re
```

```python
from collections import defaultdict
```

```python
def make_species_age_dict(filename):
    species_ages = defaultdict(list)

    for line in filename:
        species=line.split('","')[5]
        minage=line.split('","')[10]
        maxage=line.split('","')[11]
        if re.search(r'"species"', line):
            species_ages[species].append(minage + maxage)
    return species_ages
```

```python
make_species_age_dict(canid_recs)
```
