Lets make a dictionary to have a collection of all the fossils and their mean_ages; a list of every average age for that species 

```python
from collections import defaultdict #use default dict
```

```python
species_ranges = defaultdict(list) #values will be initialized as a list in the dict
#if the key is already present, it will append to the list, if its new key it will create a new key
for line in canid_recs:
    species=line.split(",")[5]
    minage=line.split('","')[10]
    maxage=line.split('","')[11] #add ' to be splitting on "info" character
    
    mean_age = (float(minage) + float(maxage)) /2 #we need to get rid of ""
    species_ranges[species].append(mean_age) #add species as keys and mean_ages as elements to the list of ages
    
```

```python
#species_ranges.values() #we can see that we have the species ranges
```

Get the oldest and youngest ages of fossils for each species

```python
output=open("canid-ranges.csv", "w") #i am making the output file

for key in species_ranges.keys():
    ages=species_ranges[key]
    minage=min(ages)
    maxage=max(ages)
    #print(key, minage, maxage)
    #now we need to get the genus
    genus=key.split(" ")[0]
    #print(genus,key, minage, maxage)
    outline= "{},{},{},{}\n".format(genus, species, minage, maxage)
    output.write(outline)
    print(outline)
```
